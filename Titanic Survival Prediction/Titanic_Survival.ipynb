<b> How to convent into numerical data?

<b>Label encoding is a technique used in machine learning to convert categorical data into numerical values so that algorithms can process it. In label encoding, each unique category value is assigned an integer number.

Label encoding is a technique used in machine learning to convert categorical data into numerical values so that algorithms can process it. In label encoding, each unique category value is assigned an integer number.

Here's a more detailed explanation:

### Why Use Label Encoding?

Machine learning algorithms typically require numerical input and output variables. Categorical data, which represent discrete values or categories (such as "red," "blue," "green"), need to be transformed into a numerical format for the algorithm to make sense of it. Label encoding is one of the simplest ways to perform this transformation.

### How Label Encoding Works

1. **Identify the categorical feature**: Determine which feature(s) in your dataset contain categorical data.
2. **Assign an integer to each unique category**: Each unique category value is assigned a distinct integer. For example, for a "Color" feature with categories ["red", "blue", "green"], the label encoding might be:
    - "red" -> 0
    - "blue" -> 1
    - "green" -> 2

### Example

Consider a dataset with a single categorical feature, "Color":

| Color |
|-------|
| red   |
| blue  |
| green |
| blue  |
| red   |

After label encoding, the dataset would look like this:

| Color |
|-------|
| 0     |
| 1     |
| 2     |
| 1     |
| 0     |

### Advantages

- **Simplicity**: It is straightforward to implement and understand.
- **Efficiency**: Uses minimal computational resources.

### Disadvantages

- **Ordinal Nature**: Label encoding imposes an ordinal relationship between the categories, which might not always be appropriate. For instance, in the example above, it implies that "red" < "blue" < "green," which might not be meaningful.
- **Non-uniqueness**: Some machine learning algorithms can misinterpret the encoded values as having a specific order or magnitude, which can lead to inaccurate results.

### Alternatives

To avoid the ordinal implications of label encoding, other encoding methods such as **one-hot encoding** can be used, where each category is represented by a binary vector.

### Implementation in Python

Hereâ€™s a basic implementation using `sklearn`:

```python
from sklearn.preprocessing import LabelEncoder

# Example data
data = ['red', 'blue', 'green', 'blue', 'red']

# Initialize label encoder
label_encoder = LabelEncoder()

# Fit and transform the data
encoded_data = label_encoder.fit_transform(data)

print(encoded_data)
```

This will output:

```
[0 1 2 1 0]
```

Label encoding is a useful preprocessing step for many machine learning tasks, but it's important to understand when it's appropriate and when other methods might be more suitable.</b>

<h3>The iloc function in pandas is used to select data by position or index. It allows you to access a group of rows and columns in a DataFrame using integer-based indexing.</h3>

````python
X = Titanic_Survival.iloc[:, 1:].values
y = Titanic_Survival.iloc[:, 0].values
````

<h5>scikit-learn, often abbreviated as sklearn, is a widely-used open-source machine learning library in Python.<br> It provides a simple and efficient set of tools for data mining and data analysis. Here's a short overview:</h5> <br>
<h3>Research and Development: </h3><br>

````python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
````

````python
from sklearn.tree import DecisionTreeClassifier
classifier = DecisionTreeClassifier()
````

````python
classifier.fit(X_train, y_train)
````

````python
y_predict = classifier.predict(X_test)
````

<h1>Finally calculating accuracy Scores<h1>

````python
from sklearn.metrics import accuracy_score
accuracy_score(y_predict, y_test)
````
